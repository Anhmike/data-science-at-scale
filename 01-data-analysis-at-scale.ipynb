### Massage data types 

Before we convert to parquet format, let's clean up our types a little

````python
%%time

#import pandas as pd

df = pd.read_csv(
    "data_taxi/yellow_tripdata_2019-01.csv", 
    parse_dates=["tpep_pickup_datetime", "tpep_dropoff_datetime"],
)
df
````

````python
df = df.astype({
    "VendorID": "uint8",
    "passenger_count": "uint8",
    "RatecodeID": "uint8",
    "store_and_fwd_flag": "category",
    "PULocationID": "uint16",
    "DOLocationID": "uint16",    
})

df["tip_amount"] = df.total_amount / df.tip_amount
````

## Convert to Parquet

````python
%%time
df.to_parquet("data_taxi/yellow_tripdata_2019-01.parq")
````

````python
%%time
df = pd.read_parquet("data_taxi/yellow_tripdata_2019-01.parq", columns=["passenger_count"])
````

## Use Dask locally to process the full dataset

````python
from dask.distributed import LocalCluster, Client
cluster = LocalCluster()
client = Client(cluster)
client
````

````python
%%time

import dask.dataframe as dd

df = dd.read_csv(
    "data_taxi/yellow_tripdata_2019-*.csv", 
    parse_dates=["tpep_pickup_datetime", "tpep_dropoff_datetime"],
    dtype={'RatecodeID': 'float64',
       'VendorID': 'float64',
       'passenger_count': 'float64',
       'payment_type': 'float64'}

)
df
````

````python
df = df.astype({
    "VendorID": "UInt8",
    "passenger_count": "UInt8",
    "RatecodeID": "UInt8",
    "store_and_fwd_flag": "category",
    "PULocationID": "UInt16",
    "DOLocationID": "UInt16",    
})

df["tip_amount"] = df.total_amount / df.tip_amount
````

````python
%%time
df.to_parquet("data_taxi/yellow_tripdata_2019.parq")
````

````python
from dask.distributed import LocalCluster, Client
````

````python
#import coiled

cluster = coiled.Cluster(n_workers=10, configuration="parquet")
````

````python
client = Client(cluster)
client
````

````python
import dask.dataframe as dd
df = dd.read_csv(
    "s3://nyc-tlc/trip data/yellow_tripdata_2019-*.csv", 
    parse_dates=["tpep_pickup_datetime", "tpep_dropoff_datetime"],
    dtype={
        'RatecodeID': 'float64',
       'VendorID': 'float64',
       'passenger_count': 'float64',
       'payment_type': 'float64'
    },
    storage_options={"anon":True}
)
df
````

````python
df = df.astype({
    "VendorID": "UInt8",
    "passenger_count": "UInt8",
    "RatecodeID": "UInt8",
    "store_and_fwd_flag": "category",
    "PULocationID": "UInt16",
    "DOLocationID": "UInt16",    
})

df["tip_amount"] = df.total_amount / df.tip_amount
````

````python
%%time
df.to_parquet("s3://coiled-data/nyctaxi-2019.parq")
````

````python
cluster.close()
````
